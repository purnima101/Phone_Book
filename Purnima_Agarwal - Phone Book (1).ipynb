```python
# phone book using python

def save():
    #function to save a new contact
    num=input("Enter a number: \n")
    if num=="" or num.isnumeric()==False:
        print("Enter a valid value:")
        return
    name=input("enter the name: \n")
    if name=="" or name.isnumeric()==True:
        print("Enter a valid value:")
        return
    num=int(num)
    phn[name]=num
    print("Entry Done!") 
    print(f"{name} : {num}")

    
def search():
    n=input("Enter the name to be searched: ")
    if n=="" or n.isnumeric()==True:
        print("Enter a valid value:")
        return

    if n in phn:
        print(phn[n])
        return
    else:
        print("Contact doesnt exists!!")
        

def delete():
    num=input("Enter the contact name to delete: ")
    if num=="" or num.isnumeric()==True:
        print("Enter a valid value:")
        return
    if num in phn:
        phn.pop(num, None)
        print("Contact deleted!!")
    else:
        print("No such contact!!")

def display():
    print("\n ------------------------------------- \n")
    for key in phn:
        
        print(f"{key} : {phn[key]}")
        
        
def update():
    print("1. Update the number \n2. Update the name")
    n=input("Enter choice: ")
    if n == '1':
        name=input("Enter the name: ")
        if name in phn:
            num=int(input("Enter new number: "))
            phn[name]=num
            print("Contact updated")
            print(f"{name} : {num}")

        else:
            print("contact doesnt exists")
    elif n== '2':
        name=input("Enter the name: ")
        if name in phn:
            name2=input("Enter new name: ")
            phn[name2]=phn.pop(name)
            print("Contact updated")
            print(f"{name2} : {phn[name2]}")
      
        else:
            print("contact doesnt exists")
    else:
        print("wrong input")
        
            

phn={}

while(1):
    #loop tp execute till user exits
    print("\n ------------------------------------- \n")
    print("Enter:\n1. Save a number\n2. Search a number\n3. Delete a number\n4. Display the phone book\n5. Update\n6. Exit.")
    print("\n ------------------------------------- \n")
    choice=input("Enter choice: ")
    if choice=="":
        #to check that no empty string is passed
        print("Enter a valid value!! \n")
        continue
   
    if choice=='1':
        save()
        continue
        
    elif choice=='2':
        search()
        continue  

    elif choice=='3':
        delete()
        continue
    
    elif choice=='4':
        display()
        continue
        
    elif choice=='5':
        update()
        continue
            
    elif choice=='6':
        print("Bye!")
        break
    
    else:
        print("Wrong Input!!!")
        continue
        
```
