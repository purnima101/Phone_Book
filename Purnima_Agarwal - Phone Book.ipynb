```python
def save():
    num=input("Enter a number: \n")
    if num=="" or num.isnumeric()==False:
        print("Enter a valid value:")
        return
    name=input("enter the name: \n")
    if name=="" or name.isnumeric()==True:
        print("Enter a valid value:")
        return
    num=int(num)
    phn[name]=num
    print("Entry Done!") 
    print(f"{name} : {num}")

    
def search():
    n=input("Enter the name to be searched: ")
    if n=="" or n.isnumeric()==True:
        print("Enter a valid value:")
        return

    if n in phn:
        print(phn[n])
        return
    else:
        print("Contact doesnt exists!!")
        

def delete():
    num=input("Enter the contact name to delete: ")
    if num=="" or num.isnumeric()==True:
        print("Enter a valid value:")
        return
    if num in phn:
        phn.pop(num, None)
        print("Contact deleted!!")
    else:
        print("No such contact!!")

def display():
    print("\n ------------------------------------- \n")
    for key in phn:
        
        print(f"{key} : {phn[key]}")
            

phn={}
count=0
while(1):
    print("\n ------------------------------------- \n")
    print("Enter 1 to enter a number, 2 to search a number, 3 to delete a number, 4 to display the phn book, 5 to exit.")
    print("\n ------------------------------------- \n")
    choice=input("Enter choice: ")
    if choice=="":
        print("Enter a valid value!! \n")
        continue
   
    if choice=='1':
        save()
        continue
        
        
    elif choice=='2':
        search()
        continue  

    elif choice=='3':
        delete()
        continue
    
    elif choice=='4':
        display()
        continue
            
    elif choice=='5':
        print("Bye!")
        break
    
    else:
        print("Wrong Input!!!")
        continue
        
```
