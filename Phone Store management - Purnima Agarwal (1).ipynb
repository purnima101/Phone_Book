```python
class Phone:
    phn={"Name":["Brand name","Price","Unit"]}
    def __init__(self,brand,name,price,unit):
        self.name=name
        self.price=price
        self.unit=unit
        self.brand=brand
        Phone.phn[self.name]=[self.brand,self.price,self.unit]
        
    def display():
        for key in Phone.phn:
            print(f"{key}  {Phone.phn[key][0]}  {Phone.phn[key][1]}  {Phone.phn[key][2]}")
        

class Store(Phone):
    
    def addPhone():
        name=input("Enter name of phone: ")
        brand=input("Enter brand")
        
        if name in Phone.phn:
            print("restocking....")
            price=int(input("Enter price: "))
            unit=int(input("Enter units: ")) + Phone.phn[name][2]
            
            Phone(brand,name,price,unit)
        else:   
        
            price=int(input("Enter price: "))
            unit=int(input("Enter units: "))
            Phone(brand,name,price,unit)

    def remove_stock(name):
        if name in Phone.phn:
            print("We have the stock")
            c=input("Do u want to remove Y/N")
            if c=="Y":
                Phone.phn[name][2]=Phone.phn[name][2]-1
                if Phone.phn[name][2]<1:
                    Phone.phn.pop(name)
                print("Phone Removed!")
            elif c=="N":
                print("ok!")
        else:
            print("Not in stock!!")
            
    def print_by_brand(brand):
        
        for (k,v) in Phone.phn.items():
            if brand in v:
                print(f"{k}  {v[0]}  {v[1]}  {v[2]}")
                
            
        


while True:
    
    print("---------------------------------------")
    print("\n1. Add New Phone \n2. Remove Stock \n3. Display all phones \n4. Exit \n5. Search by brand ")
    print("---------------------------------------")
    choice=input()
    if choice.isnumeric()== False or choice =="":
        print("Enter correct values")
        continue
    elif choice=='1':
        Store.addPhone()
        continue
    elif choice=='2':
        n=input("Enter name of phone to remove: ")
        Store.remove_stock(n)
    elif choice=='3':
        Store.display()
    elif choice=='4':
        break
    elif choice=='5':
        n=input("Enter brand : ")
        Store.print_by_brand(n)
            
    else:
        continue
        
        
```
