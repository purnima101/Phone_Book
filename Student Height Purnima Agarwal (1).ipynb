```python
class Height:
    details={}
    
    def __init__(self,name,cls,section,roll,height):
        self.name=name
        self.height=height
        self.cls=cls
        self.section=section
        self.roll=roll
        Height.details[self.name]=[self.cls,self.section,self.roll,self.height]
        
    def display():
        for key in Height.details:
            print(f"{key} : {Height.details[key][0]}  {Height.details[key][1]}  {Height.details[key][2]}  {Height.details[key][3]}")
    
    def search(name):
        
        if name in Height.details:
            print(f"{name} : {Height.details[name][0]}  {Height.details[name][1]}  {Height.details[name][2]}  {Height.details[name][3]}")
        else:
            print("\nName not in records!")

    def delete(name):
       
        if name in Height.details:
            Height.details.pop(name)
            print("\nRecord deleted!!")
        else:
            print("Enter a valid name!!!")
            
    def modify(name):
        if name in Height.details:
            height=input("enter new height: ")
            Height.details[name][3]=float(height)
            print("\nRecord modified!!")
        else:
            print("Enter a valid name!!!")
            
        
            
    
    
        
while True:
    print("\n------------------------------------\n")
    print("1. Record height \n2. Display all records \n3. Search by name \n4. Delete a record \n5. Modify record \n6. Exit \n")
    print("\n------------------------------------\n")
    
    choice=input("Enter your choice: ")
    
    if choice=="" or choice.isnumeric()==False:
        print("\nEnter a valid choice!")
        continue
                  
    if choice=='1':
        name=input("Enter NAME: ")
        if name=="":
            print("\nEnter VALID NAME! ")
            continue
        cls=input("Enter class: ")
        if cls=="":
            print("\nEnter VALID class! ")
            continue
        section=input("Enter Section: ")
        if section=="":
            print("\nEnter VALID section! ")
            continue
        roll=input("Enter Roll: ")
        if roll=="":
            print("\nEnter VALID Roll! ")
            continue
        height=input("Enter height in centimeter: ")
        if height=="":
            print("\nEnter VALID height! ")
            continue
        
        Height(name,cls,section,roll,float(height))
        print("\nRecord Inserted!!!")
        continue
                  
    elif choice=='2':
        Height.display()
        continue
                  
    elif choice=='3':
        name=input("Enter the name: ")
        
        Height.search(name)
        continue
    
    elif choice=='4':
        name=input("Enter the name: ")
        
        Height.delete(name)
        continue
    
    elif choice=='5':
        name=input("Enter the name: ")
        
        Height.modify(name)
        continue
                     
    elif choice=='6':
        break
    
    else:
        continue
        
        
```
